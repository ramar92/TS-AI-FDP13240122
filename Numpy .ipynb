```python
import numpy as np 
arr = ([3,4,6],[5,6,7],[6,56,7]) 
print(np.amin(arr))
print(np.amax(arr))


print(np.mean(arr))
print(np.median(arr)) 
print(np.std(arr))
print(np.var(arr))
print(np.ptp(arr)) 
print(np.average(arr))
```

```python
# Precentile & Interquartile using numpy 

# IQR = Q3 - Q1  

import numpy as rit 

data = [34,56,78,32,44,6,8,90,66,44,67,99,88,44,55] 

#first quartile (Q1)

Q1 = rit.median(data[:7]) 

#thir quatrial(Q3)
Q3 = rit.median(data[7:]) 

#interquartile range(IQR)
IQR = Q3 - Q1 
print(IQR)
```
