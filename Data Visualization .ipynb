```python
import matplotlib.pyplot as plt  

# x axis values (temperature)
temp_x = [10,12,33,24,15,16,27]

#y axis Values (Huminity)
hum_y = [2,4,6,8,10,12,14] 

#plotting points 
plt.plot(temp_x, hum_y) 

# lable name 
plt.xlabel("X - Temp")
plt.ylabel("Y - Hum")

#title 
plt.title("My First Graph")

#show first graph
plt.show()
```

```python
## Bar plot 

import matplotlib.pyplot as plt  

# x axis values (temperature)
temp_x = [10,12,33,24,15,16,27]

#y axis Values (Huminity)
hum_y = [2,4,6,8,10,12,14] 


#funtion to plot 

plt.bar(temp_x,hum_y)
# lable name 
plt.xlabel("X - Temp")
plt.ylabel("Y - Hum")

#title 
plt.title("My Bar Graph")
plt.show()
```

```python
## Scatter plot 


import matplotlib.pyplot as plt  

# x axis values (temperature)
temp_x = [10,12,33,24,15,16,27]

#y axis Values (Huminity)
hum_y = [2,4,6,8,10,12,14] 


#function to plot the scatter 

plt.scatter(temp_x, hum_y)

# lable name 
plt.xlabel("X - Temp")
plt.ylabel("Y - Hum")

#title 
plt.title("My Scatter Graph")
plt.show()
```

```python
#plotting 3d lines and points  


import numpy as np 
import matplotlib.pyplot as plt 

fig = plt.figure()
ax = plt.axes(projection = '3d') 


#define data for all 3 axes 
z = np.linspace(0,1,100) 
x = z * np.sin(25 * z)
y = z * np.cos(25 * z) 

# plotting 
ax.plot3D(x,y,z, 'green')
ax.set_title('3D line plot')
plt.show()
```

```python
#Pie plot chart using matplotlib 

#import the lib 
import numpy as np 
from matplotlib import pyplot as plt 

#creating data set 
cat_car = ['Audi','Bmw','Ford','Maruti','hundai','Tesla'] 
car_sales = [3000,2500,5000,10000,9000,30] 

#create the plot 
fig = plt.figure(figsize=(10,7))
plt.pie(car_sales, labels = cat_car)

plt.show()
```
